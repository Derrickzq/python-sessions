How to update matplotlib's default parameter for more readable plots and better math rendering.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n_points = 1000
y = 0.1 + 1. / np.arange(1, n_points + 1)
```

```python
def dummy_plot():
    plt.figure()
    plt.semilogy(y)
    plt.xlabel("Iteration $k$")
    plt.ylabel('$f(x_k) - f(x^*)$')
    plt.title('A non converging algorithm')
    plt.show()
    
dummy_plot()
```

You can update matplotlib's global parameter to have bigger (more readable) fonts by default, and also to use LaTeX for rendering of math equations:

```python
# choose a large font size by default and use tex for math
fontsize = 18
params = {'axes.labelsize': fontsize + 2,
      'font.size': fontsize + 2,
      'legend.fontsize': fontsize + 2,
      'xtick.labelsize': fontsize,
      'ytick.labelsize': fontsize,
      'text.usetex': True}
plt.rcParams.update(params)
```

```python
dummy_plot()
```

You can also change the font to match the font of your paper/report (provided the font is installed on your machine and matplotlib finds it, which can be tricky):

```python
plt.rc('font', family='Times New Roman')
```

```python
dummy_plot()  # title and xlabel font have changed
```
